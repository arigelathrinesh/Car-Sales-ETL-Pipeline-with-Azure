#Creating Fact

**Querying silver data**

```python
df_silver = spark.sql("select * from cata.silver.transformed_tbl")
```

```python
df_silver.display()
```

**Reading all dimensions**

```python
df_branch = spark.sql('select * from cata.gold.dim_branch')
df_date = spark.sql('select * from cata.gold.dim_date')
df_dealer = spark.sql('select * from cata.gold.dim_dealer')
df_model = spark.sql('select * from cata.gold.dim_model')
```

**Joining the Fact and Dimensiomns**

```python
df_fact =  df_silver.join(df_branch, df_silver.Branch_ID == df_branch.Branch_ID, 'left')\
                    .join(df_date, df_silver.Date_ID == df_date.Date_ID, 'left')\
                    .join(df_dealer, df_silver.Dealer_ID == df_dealer.Dealer_ID, 'left')\
                    .join(df_model, df_silver.Model_ID == df_model.Model_ID, 'left')\
                    .select(df_silver['Revenue'],df_silver['Units_Sold'],df_silver['Revenue/Unit'],df_branch['dim_branch_key'],df_date['dim_date_key'],df_dealer['dim_dealer_key'],df_model['dim_model_key'])
```

```python
df_fact.display()
```

#Writing The Fact Table

```python
from delta.tables import DeltaTable
```

```python
if spark.catalog.tableExists('cata.gold.fact_table'):
    delta_tbl = DeltaTable.forPath(spark, 'abfss://gold@thrineshproject1.dfs.core.windows.net/fact_table')

    delta_tbl.alias('tar')\
    .merge(df_fact.alias('src'), 'tar.dim_branch_key = src.dim_branch_key AND tar.dim_date_key = src.dim_date_key AND tar.dim_dealer_key = src.dim_dealer_key AND tar.dim_model_key = src.dim_model_key')\
    .whenMatchedUpdateAll()\
    .whenNotMatchedInsertAll()\
    .execute()

else:
    df_fact.write.mode('overwrite')\
           .format('delta')\
           .option('path','abfss://gold@thrineshproject1.dfs.core.windows.net/fact_table')\
           .saveAsTable('cata.gold.fact_table')
```
